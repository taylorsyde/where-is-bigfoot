```python
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
states_file = "Resources/csv_states.csv"
states_df = pd.read_csv(states_file)
states_df.head()
states_df = states_df.rename(columns={"State": "state"})
states_df
```

```python
bigfoot_file = "Resources/bfro_reports_geocoded.csv"
bigfoot_df = pd.read_csv(bigfoot_file)
#bigfoot_df.head()
```

```python
bigfoot_clean_df = bigfoot_df[['number','date','county', 'state','latitude', 'longitude', 'observed']]
bigfoot_clean_df.head(10)
bigfoot_clean_df2 = bigfoot_clean_df.fillna(0)
bigfoot_clean_df2.head()
```

```python
#merging bigfoot with states 
#bigfoot_merge = pd.merge(bigfoot_clean_df, states_df, on="state")
bigfoot_merge = pd.merge(bigfoot_clean_df2, states_df, on="state", how="left")
bigfoot_merge.head()
#bigfoot_merge['state'].isnull().values.any()
big_foot_drop2 = bigfoot_merge.drop(columns=['state', 'Abbrev'])
#big_foot_drop2.head()
bigfoot_complete = big_foot_drop2.rename(columns={"Code": "state"})
bigfoot_complete.head()

bigfoot_complete["date"] = pd.to_datetime(bigfoot_complete['date']).dt.date
print(type(bigfoot_complete['date'][1]))
bigfoot_complete.head()
```

```python
ufo_file = "Resources/ufo_sightings.csv"
ufo_df = pd.read_csv(ufo_file)
ufo_df.head(10)
```

```python
# df['expiry_date'] = pd.to_datetime(df['expiry_date'], errors='coerce').dt.strftime('%Y-%m-%d')
# print(df)

ufo_clean_df = ufo_df[['datetime', 'city', 'state', 'country', 'latitude', 'longitude', 'comments']]
ufo_clean_df.head(10)
ufo_clean_df['state'] = ufo_clean_df['state'].str.upper()

#ufo_clean_df['datetime'].isnull().values.any()
ufo_clean_df
#ufo_clean_df["datetime"] = pd.to_datetime(ufo_clean_df["datetime"]).apply(lambda x:x.strftime("%Y-%m-%d"))

#ufo_clean_df["datetime"] = pd.to_datetime(ufo_clean_df["datetime"],format='%d%b%Y', errors='ignore')


ufo_clean_df["datetime"] = pd.to_datetime(ufo_clean_df['datetime'].str.replace('/', '-'), errors='coerce').dt.date

print(type(ufo_clean_df['datetime'][0]))
#ufo_clean_df["datetime"] = pd.to_datetime(ufo_clean_df["datetime"]).apply(lambda x:x.strftime("%Y-%m-%d"))
ufo_clean_df
ufo_clean_df.dropna()

ufo_final = ufo_clean_df.rename(columns={"datetime": "date"})

# drop rows with lat = 0
ufo_final = ufo_final.loc[(ufo_final['longitude'] != 0),:]
ufo_final = ufo_final.loc[(ufo_final['latitude'] != '33q.200088'),:]

# change lat to float
ufo_final["latitude"] = ufo_final["latitude"].astype("float")
ufo_final
```
